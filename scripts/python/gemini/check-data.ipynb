```python
# replace -1 in num_reviews with None
isbndb_goodreads['num_reviews'] = isbndb_goodreads['num_reviews'].apply(lambda x: None if x == -1 else x)
isbndb_goodreads['num_ratings'] = isbndb_goodreads['num_ratings'].apply(lambda x: None if x == -1 else x)
isbndb_goodreads['goodreads_year'] = isbndb_goodreads['goodreads_year'].apply(lambda x: None if x == -1 else x)

print(isbndb_goodreads.isnull().sum())
```

```python
# only keep books with goodreads_year not None
isbndb_goodreads = isbndb_goodreads[isbndb_goodreads['goodreads_year'].notna()]
print(isbndb_goodreads.shape)
print(isbndb_goodreads.isnull().sum())
```
